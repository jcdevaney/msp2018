Import libraries and give them abbreviations

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

Open entire piano audio file as mono and store the original sampling frequency in a variable

```python
file = 'r0.wav'
sig , sr = librosa.load(file,mono=True,sr=None)
```

Calculate the length in seconds 

```python
length_s = sig.shape[0] / sr
print(length_s)
```

Plot signal

```python
plt.plot(sig)
```

Plot x2 with respect to time

```python
time=np.arange(0,sig.shape[0])/sr
```

```python
plt.plot(time,sig)
```

Load and plot violin and voice audio files

```python
violinFile = 'h.wav'
voiceFile = 'mc.wav'

violinSig , sr2 = librosa.load(violinFile,mono=True,sr=None)
voiceSig , sr3 = librosa.load(voiceFile,mono=True,sr=None)

timeViolin=np.arange(0,violinSig.shape[0])/sr2
timeVoice=np.arange(0,voiceSig.shape[0])/sr3
```

Plot time-domain signal before resampling

```python
plt.plot(timeViolin,violinSig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Violin Waveform')
```

```python
plt.plot(timeVoice,voiceSig)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Voice Waveform')
```

Plot linear spectrograms

```python
violinSpec = librosa.amplitude_to_db(np.abs(librosa.stft(violinSig)), ref=np.max)
```

```python
librosa.display.specshow(violinSpec, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('Violin: Linear-frequency power spectrogram')
```

```python
vocalSpec = librosa.amplitude_to_db(np.abs(librosa.stft(voiceSig)), ref=np.max)
```

```python
librosa.display.specshow(vocalSpec, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('Vocal: Linear-frequency power spectrogram')
```

Plot logarithmic spectrograms

```python
librosa.display.specshow(violinSpec, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Violin: Log-frequency power spectrogram')
```

```python
librosa.display.specshow(vocalSpec, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Vocal: Log-frequency power spectrogram')
```
